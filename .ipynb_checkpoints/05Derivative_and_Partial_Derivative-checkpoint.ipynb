1.Derivative（求导），精髓在于“变化率” <br>
2.Partial Derivative（偏导数） ,推荐用sympy库求偏导并显示公式<br>
multivariable function <br>

```python
import sympy as sym
sym.init_printing()
x,y = sym.symbols("x y")
f = x**2 * y**2 + sym.cos(y) + sym.sin(x)
```

```python
sym.Derivative(f,x)
```

```python
sym.diff(f,x)
```

```python
sym.Derivative(f,y)
```

```python
sym.diff(f,y)
```

```python
%matplotlib inline

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection = "3d")
```

```python
%matplotlib notebook

def func(x,y):
    return x**2 * y**2 + np.cos(y) + np.sin(x)

x = np.linspace(-888,888,8)
y = np.linspace(-888,888,8)
```

```python
x
```

```python
X, Y = np.meshgrid(x,y)
```

```python
X
```

```python
Z = func(X,Y)
Z
```

```python
fig = plt.figure()
ax = plt.axes(projection = "3d")
ax.contour3D(X,Y,Z,5,cmap="binary")
```
