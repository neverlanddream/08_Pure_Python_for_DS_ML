```python
import sympy as sym
sym.init_printing()
x,y = sym.symbols("x y")
f = x**2 * y**2 + sym.cos(y) + sym.sin(x)
```

```python
sym.Derivative(f,x)
```

```python
sym.diff(f,x)
```

```python
sym.Derivative(f,y)
```

```python
sym.diff(f,y)
```

```python
%matplotlib notebook

def func(x,y):
    return x**2 * y + np.cos(y) + np.sin(x)

x = np.linspace(-888,888,8)
y = np.linspace(-888,888,8)
```

```python
x
```

```python
X, Y = np.meshgrid(x,y)
```

```python
X
```

```python
Z = func(X,Y)
Z
```

```python
fig = plt.figure()
ax = plt.axes(projection = "3d")
ax.contour3D(X,Y,Z,5,cmap="binary")
```

```python
ax = plt.axes(projection = "3d")
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap="viridis",edgecolor="none")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
```

梯度Gredient<br>
$function : f(x,y)$ <br>
$Gredient \: Vecter : \nabla f(x,y) = \begin{bmatrix}
\frac{\partial f(x,y)}{\partial x} \\
\frac{\partial f(x,y)}{\partial y}
\end{bmatrix} = \begin{bmatrix} f_x \\ f_y \end{bmatrix} $<br>
梯度就是x方向的向量和y方向的向量的和，此向量和刚好就是这个点变化率最快的向量。

散度Divergence <br>
梯度是多元函数的一次偏导的矢量和，表示函数的变化方向及其大小，代表函数的变化方向与大小。 <br>
散度是多元函数二次偏导的代数和，表示函数梯度的发散的强弱程度，代表函数梯度矢量场Vector Field的发散的强度。 <br>
因此，对梯度这一矢量再做求梯度运算得到的是标量，即散度。 <br>
